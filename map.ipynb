```python
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
import numpy as np

import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
import numpy as np

def plot_wind_vectors(df, date):
    # Crea un GeoDataFrame a partir del DataFrame original
    gdf = gpd.GeoDataFrame(df, 
                           geometry=df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1), 
                           crs="EPSG:4326")

    # Filtra los datos para la fecha especificada
    df_filtered = gdf[gdf['Date'] == date]

    if df_filtered.empty:
        print(f"No hay datos disponibles para la fecha {date}.")
        return

    # Obtiene todas las direcciones únicas del viento presentes en los datos filtrados
    unique_wind_directions = df_filtered['WindGustDir'].dropna().unique()

    # Crea un gráfico del mapa de Australia
    fig, ax = plt.subplots(figsize=(15, 15))
    map_df.plot(ax=ax, alpha=0.4, color='grey')

    # Dibuja vectores de viento en las ubicaciones de las ciudades para cada dirección única
    for wind_dir in unique_wind_directions:
        wind_dir_df = df_filtered[df_filtered['WindGustDir'] == wind_dir]

        for index, row in wind_dir_df.iterrows():
            city = row['Location']
            lat = row['Latitude']
            lon = row['Longitude']
            wind_speed = row['WindGustSpeed']

            if not np.isnan(wind_speed):
                # Convierte la dirección del viento en un ángulo en radianes
                wind_directions = unique_wind_directions
                angle = (wind_directions.tolist().index(wind_dir) * (360/len(unique_wind_directions))) * (np.pi / 180)

                # Calcula los componentes x e y del vector de viento
                u = wind_speed * np.cos(angle)
                v = wind_speed * np.sin(angle)

                # Dibuja el vector de viento en la ubicación de la ciudad con una escala más pequeña
                ax.quiver(lon, lat, u, v, angles='xy', scale_units='xy', scale=20, color='blue', alpha=0.7)

    # Añade un título y muestra el gráfico
    plt.title(f'Campo Vectorial de Dirección del Viento en Australia - Fecha: {date}', fontsize=16)
    plt.show()

# Ejemplo de uso:
plot_wind_vectors(gdf, '2008-12-01')
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
import numpy as np

def calculate_wind_vectors(df, date):
    # Crea un GeoDataFrame a partir del DataFrame original
    gdf = gpd.GeoDataFrame(df, 
                           geometry=df.apply(lambda row: Point(row['Longitude'], row['Latitude']), axis=1), 
                           crs="EPSG:4326")

    # Filtra los datos para la fecha especificada
    df_filtered = gdf[gdf['Date'] == date]

    if df_filtered.empty:
        print(f"No hay datos disponibles para la fecha {date}.")
        return

    # Obtiene todas las direcciones únicas del viento presentes en los datos filtrados
    unique_wind_directions = df_filtered['WindGustDir'].dropna().unique()

    # Crea un gráfico del mapa de Australia
    fig, ax = plt.subplots(figsize=(15, 15))
    map_df.plot(ax=ax, alpha=0.4, color='grey')

    # Define el espaciado de la malla de puntos
    spacing = 1.0  # Ajusta el espaciado según tus preferencias

    # Crea una malla de puntos sobre el mapa de Australia
    x = np.arange(map_df.bounds.minx.min(), map_df.bounds.maxx.max(), spacing)
    y = np.arange(map_df.bounds.miny.min(), map_df.bounds.maxy.max(), spacing)
    xx, yy = np.meshgrid(x, y)

    # Dibuja vectores de viento en la malla de puntos
    for wind_dir in unique_wind_directions:
        for i in range(len(x)):
            for j in range(len(y)):
                lon = xx[j, i]
                lat = yy[j, i]

                # Calcula la velocidad del viento para cada punto de la malla
                # Puedes calcularla de manera ficticia o usar datos reales si los tienes
                wind_speed = np.random.uniform(0, 20)  # Ejemplo de velocidad ficticia

                # Convierte la dirección del viento en un ángulo en radianes
                wind_directions = unique_wind_directions
                angle = (wind_directions.tolist().index(wind_dir) * (360/len(unique_wind_directions))) * (np.pi / 180)

                # Calcula los componentes x e y del vector de viento
                u = wind_speed * np.cos(angle)
                v = wind_speed * np.sin(angle)

                # Dibuja el vector de viento en el punto de la malla
                ax.quiver(lon, lat, u, v, angles='xy', scale_units='xy', scale=20, color='blue', alpha=0.7)

    # Añade un título y muestra el gráfico
    plt.title(f'Campo Vectorial de Dirección del Viento en Australia - Fecha: {date}', fontsize=16)
    plt.show()

# Ejemplo de uso:
calculate_wind_vectors(gdf, '2008-12-01')
```
